### **Regularização `Ridge`**
---
---

Quando aplicamos um _Regressão Linear_ usando a regularização `Ridge (L2)`, os coeficientes (pesos) serão forçados a ser os menores possíveis, enquanto o algoritmo se ajusta aos dados.

Quando definimos o hiperparâmetro _alfa_ mais próximo de zero, isso significa que queremos menos regularização, então o modelo terá pesos mais altos e a linha terá uma inclinação maior. Por outro lado, se definirmos _alfa_ para um valor mais alto, os pesos serão forçados a serem menores, resultando em números mais baixos e uma linha mais plana, mais próxima do valor médio do conjunto de dados.

```python
import numpy as np
import scipy.stats as scs
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from sklearn.linear_model import LinearRegression, Ridge
```

```python
%matplotlib inline
# plt.style.use('ggplot')
sns.set()
warnings.filterwarnings('ignore')
```

```python
X = scs.norm.rvs(0, 0.1, size=100) - 20
y = X * scs.norm.rvs(1, 40, size=100) + 0.5
```

```python
type(X), type(y)
```

```python
X[:20]
```

```python
y[:20]
```

```python
plt.figure(figsize=(20, 7))

dummy_ridge = Ridge(alpha=0.001).fit(X.reshape(-1, 1), y)
preds = dummy_ridge.predict(X.reshape(-1, 1))
plt.subplot(121)
sns.scatterplot(x=X, y=y, alpha=0.8)
sns.lineplot(x=X, y=preds, color='red')
plt.title(f'alpha = 0.001 | slope = {dummy_ridge.coef_}');

dummy_ridge = Ridge(alpha=1_000).fit(X.reshape(-1, 1), y)
preds = dummy_ridge.predict(X.reshape(-1, 1))
plt.subplot(122)
sns.scatterplot(x=X, y=y, alpha=0.8)
sns.lineplot(x=X, y=preds, color='red')
plt.title(f'alpha = 1.000 | slope = {dummy_ridge.coef_}');
```
